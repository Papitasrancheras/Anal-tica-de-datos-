```python
#Importar las librerías numpy y pyplot 
import numpy as np 
from matplotlib import pyplot as plt 

#Definimos 2 arreglos Numpy
ys = 200 + np.random.randn(100)
x = np.arange (0,100, 1)
print(x)

# Graficamos los arreglos en un plano 2D Pyplot
plt.plot(x, ys, ".")
plt.fill_between(x, ys, 195 , where=(ys>195), facecolor="r", alpha=0.7)
```
